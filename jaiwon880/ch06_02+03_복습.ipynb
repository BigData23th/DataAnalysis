```python
#@title 합성 how="", merge(left, right), join(인덱스 강제/ 중복되는 값), concat(걍 합침)
customers = ["Alice", "Ben", "Chales", "Don", "Ein"]
ids = [f"1{i+1:03d}" for i in range(len(customers))]
ids, customers
```

```python
df1 = pd.DataFrame({
    "id": ids,
    "name": customers,
})

df1
```

```python
df2 = pd.DataFrame({
    "id": np.random.choice([f"1{i+1:03d}" for i in range(len(customers)+2)], 10, replace=True),
})
df2
```

```python
df2.info()
```

```python
df2["money"] = df2.id.apply(lambda x: int(x) * abs(np.random.randn())* 1000).apply(int) # 행단위
df2
```

```python
pd.merge(df1, df2) # id별로 다른 값이 나옴
```

```python
pd.merge(df1, df2, how="inner") # 공통되는 것만
```

```python
pd.merge(df1, df2, how="outer") 
```

```python
pd.merge(df1, df2, how="left")
```

```python
pd.merge(df1, df2, how="right")
```
