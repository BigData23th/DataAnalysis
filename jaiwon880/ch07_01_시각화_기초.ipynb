### 그리드 설정 (grid) 격자


```python
# 보조선 활성화/비활성화 : plt.grid(T/F)
def draw_line(grid=False, style="b-"):
    plt.title("나이와 재산의 상관관계")
    plt.plot(
        range(10,50, 10), # x축 데이터
        [1,4,9,16], # y축 데이터
        style # 차트 스타일
    )
    # plt.xlabel("x축라벨")
    plt.xlabel("나이")
    plt.grid(grid)
    # plt.ylabel("y축라벨")
    plt.ylabel("재산")
    plt.show()
```

```python
draw_line(grid=True)  #인자에 True로 덮어씌움
# grid=False < 기본값
draw_line(grid=False) 
```

### 여러 개의 선 그어보기

```python
t = np.arange(0., 5., 0.2)
t
```

```python
# 1. 여러 개의 선 그리기 (하나의 plot 사용)
plt.title("라인 플롯에서 여러 개의 선")
plt.plot(
    t, t, 'r--', # 첫번째 플롯 그림
    t, t ** 2, 'bs:', 
    t, t ** 3, 'g^-', 
)

plt.show()
```

```python
# 2. 하나의 plot이 아니라 여러 개 -> 겹치는.
plt.title("여러 개의 plot 명령으로 한 그림에 겹쳐서 표현")
plt.plot((np.arange(4)+ 1) ** 2)
plt.plot((np.arange(4)- 1) ** 3)
plt.show()
```

#### `subplot`
* plot 명령을 여러 번 실행 => 같은 그래프 내부에 여러 직선 표시
* subplot : 행과 열 단위로 묶어서 여러 데이터를 병행해서 표시

```python
import numpy as np
```

```python
# 위아래 행 배치
# 행, 열, 서브플롯의 위치
plt.subplot(2, 1, 1)
plt.plot(np.arange(5) ** 2)
plt.title("차트1")
plt.subplot(2, 1, 2)
plt.plot(np.arange(10), np.arange(10) ** 3, "r--")
plt.title("차트2")
# 겹치는 문제
plt.tight_layout() # 안 겹치게 레이아웃 조정
plt.show()
```

```python
# 양옆 열 배치
# 행, 열, 서브플롯의 위치
plt.subplot(1, 2, 1)
plt.plot(np.arange(5) ** 2)
plt.title("차트1")
plt.subplot(1, 2, 2)
plt.plot(np.arange(10), np.arange(10) ** 3, "r--")
plt.title("차트2")
# 겹치는 문제
plt.tight_layout() # 안 겹치게 레이아웃 조정
plt.show()
```

```python
# 2x2 플롯 배치
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
plt.plot(np.arange(5) ** 2)
plt.title("차트1")
plt.xlabel("x축")
plt.ylabel("y축")

plt.subplot(2, 2, 2)
plt.plot(np.arange(10), np.arange(10) ** 3, "r--")
plt.title("차트2")

plt.subplot(2, 2, 3)
plt.plot(np.arange(5) ** 3)
plt.title("차트3")

plt.subplot(2, 2, 4)
plt.plot(np.arange(10), np.arange(10) ** 2, "b^-")
plt.title("차트4")
# 겹치는 문제
plt.tight_layout() # 안 겹치게 레이아웃 조정
plt.show()
```

### 다양한 차트 활용
* `lineplot` 선 차트
* 그외에 여러 가지 차트가 존재

```python
y = [2, 3, 1]
x = np.arange(3)
xlabel = ["가", "나", "다"] # len(x) 의 길이와 같아야 함
plt.title("바 차트")
plt.xticks(x, xlabel) # X축의 개별 값에다가 이름을 붙여줌
plt.yticks(sorted(y)) # y축에다가 기준이 되는 이름을 붙여주기
plt.bar(x, y)
plt.show()
```

```python
y = [2, 3, 1]
x = np.arange(3)
xlabel = ["가", "나", "다"] # len(x) 의 길이와 같아야 함
plt.title("가로 바 차트")
plt.yticks(x, xlabel) # X축의 개별 값에다가 이름을 붙여줌
plt.xticks(sorted(y)) # y축에다가 기준이 되는 이름을 붙여주기
plt.barh(x, y) # 가로 바 차트
plt.show()
```

```python
people = ["James", "John", "Kate", "Alice"]
y_position = np.arange(4)
performance = 3+10 * np.random.rand(4)
error = np.random.rand(4)


plt.title("Bar-h Cahrt")
plt.grid(True)
plt.barh(y_position, performance, xerr = error, alpha=0.4)
plt.yticks(y_position, people )

plt.xlabel("x축")
plt.show()
```

#### 파이 차트
* 원형의 파이 모양 그래프

```python
labels = ["호박파이", "사과파이", "정어리파이", "엄마손파이"]
sizes = [15, 30, 45, 10]
colors =["yellowgreen", "gold", "lightskyblue", "purple"]
explode = (0, 0.1, 0.2, 0.4)
plt.title("파이 차트")
plt.pie(sizes, explode = explode, labels =labels, colors =colors,
        autopct="%1.1f%%", shadow=True, startangle=90)

plt.axis('equal')
plt.show()
```

#### 히스토그램 (Histogram)

```python
x = np.random.randn(1000)
plt.figure(figsize=(15,5))
plt.title("히스토그램")
arrays, bins, patches = plt.hist(x, bins=10) # bins = 10개의 구간으로 묶어줌
plt.show()
```

#### 스캐터 플롯 (산점도)
