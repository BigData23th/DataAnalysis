```python
# 가격이 1만원 주식
price = 10000
# 일간 수익률 -> 기댓값(평균) 0, 표준편자 1%=0.01 (표준정규분포)
days = 250
daily_earning_rate = np.random.randn(days) / 100
print(daily_earning_rate)
# 250일 동안의 주가를 무작위로 생성
# 주가 -> 전일 가격 * (1 + 당일 수익률)
price_arr = np.empty(days)
for i, v in enumerate(daily_earning_rate):
    print(v)
    price *= (1 + v)
    print(price)
    price_arr[i] = price
    print(price_arr[i])
    print(price_arr)
price_arr
```
