## `merge` (두 개 이상의 데이터프레임을 join, 연결)

````python
import numpy as np
import pandas as pd
````

````python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1 # 고객 목록
````

````python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2 # 구매 기록
````

````python
# merge : 결합하다 -> pandas.merge(df1, df2) # df1 (left), df2 (right)
pd.merge(df1,df2)
````

````python
pd.merge(left=df2,right=df1)
````

````python
pd.merge(left=df1,right=df2,on="고객번호")
````

````python
pd.merge(left=df1,right=df2,on="고객번호",how='inner') # -> 기본 상태
# inner join:(기준되는 열) 양쪽 데이터프레임에 모두 속하는 원소들만 남겨줌
````

````python
pd.merge(df1,df2,how="inner")
````

````python
pd.merge(df1,df2,how="outer")
````

````python
pd.merge(df1,df2,how="left")
````

````python
pd.merge(df1,df2,how="right")
````

## 그룹분석
* 조건에 맞는 데이터가 하나 이상일 경우에 -> 데이터 그룹을 이룸 (피봇 테이블 형태로 표현을 X)
* 그룹의 특성을 보여주는 (통계치, 계산...) => 그룹 분석 (group analysis)
* group -> `groupby`

1. 분석하고자 하는 Series, df에 `groupby` 메소드 호출 -> 그룹화
1. 그룹화 -> 그룹 객체 -> 그룹 연산

### groupby 메소드
* 데이터를 그룹별로 분류하는 역할.
    * 열 또는 열의 리스트 (어떤 열을 기준으로 그룹을 넣어줄 것이냐
    * 행 인덱스
=> GroupBy (그룹화) => 그룹 함수 or 그룹 연산 => 특정한 값. 배열...

````python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2
````

````python
df2.info()
````

````python
groups=df2.groupby(df2["key1"])
groups
````

````python
# groupby.groups(속성) : 그룹별 구성 현황
groups.groups # 행 인덱스
````

````python
groups.count()
````

````python
df3=df2.copy()
df3.iloc[1,1]=np.nan
df3.groupby(df3.key1).count()
````

````python
groups.size()
````

````python
# 평균
groups.mean()
````

````python
# 중앙값 (median)
groups.median()
````

````python
# 최대값
groups.max()
````

````python
# 최소값
groups.min()
````

````python
# 합계
groups.sum()
````

````python
# 일괄적으로 곱함
groups.prod()
````

````python
# 표준편차, 분산, 중간값(50%)
````

````python
groups.quantile(0.75)["data2"]
````

````python
# 특정한 열만 뽑아주고 싶다
groups.sum()["data1"]
````

````python
import seaborn as sns # 예시 데이터들을 불러올 수 있음
iris = sns.load_dataset('iris') # 붓꽃
iris # 붓꽃의 데이터
````

````python
iris.head()
````

````python
iris.tail()
````

````python
iris.info()
````

````python
iris.describe(include='all')
````

````python
def peak_to_peak_ratio(x):
  return x.max()/x.min()
````

````python
# 각 붓꽃의 종별로 (열별로) 가장 큰 값과 가장 작은 값의 비율을 구해보기
iris.groupby(iris.species).agg(lambda x:x.max()/x.min())
````

````python
iris.species.unique()
````

````python
iris.groupby(iris.species).agg(lambda x:x.max()/x.min())
````

* `agg` : 그룹 들의 대표값
* `apply` : 그룹들로 나뉘어진 행에 대해서 반복적 연산

````python
def top3_petal_length(df):
  return df.sort_values(by="petal_length",ascending=False)[:3]
def top1_petal_length(df):
  return df.sort_values(by="petal_length",ascending=False)[:1]

iris.groupby(iris.species).apply(top1_petal_length)
````
