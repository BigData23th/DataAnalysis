## loc

* 문자열 인덱스 -> 인덱싱, 슬라이싱
* loc[행 인덱스], loc[행, 열 인덱스]
* 행 인덱스 하나만 넣으면 Series type으로 반환


```python
row_name ='w' #@param ['r','o','w','s']{'type':'string'}
df.loc[row_name]
```

```python
print(df.loc['r':'w'])
print('------------------')
print(df.loc['w'::-1])
```

```python
df.loc[['r','o','s']]
```

```python
df.loc[df.c>14]
```

```python
df2 = pd.DataFrame(
    np.arange(16).reshape(4,4),
    columns=list('ABCD')
)
df2
```

```python
df2.iloc[1:2]
```

```python
# [행,열] -> 모두 문자열 인덱스여야 함
# 열에도 인덱싱, 슬라이싱, 리스트
# loc의 열 인덱스 작업에서는 boolean 불가
df.loc['r','c']
```

```python
#@title 합성(merge)
customer = ['sw','bh','us','ko','jp']
ids = lambda x:[f'1{i+x:03d}' for i in range(len(customer))]
ids(1),customer
```

```python
df1 = pd.DataFrame({
    'id':ids(1),
    'name':customer
})
df1
```

```python
df2 = pd.DataFrame({
    'id' : np.random.choice(ids(2),size=10,replace=True)
})
df2
```

```python
df2['money'] = df2.id.apply(lambda x:int(x)*abs(np.random.randn())*1000).astype(int)
df2
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='inner')
```

```python
df3 = df1.copy()
df3.columns = ['id','data']
df4 = df2.copy()
df4.columns = ['id','data']

pd.merge(df3,df4,on='id')
```

```python
df5 = df1.copy()
df5.columns = ['id','data']
df6 = df2.copy()
df6.columns = ['Id','data']

pd.merge(df5,df6,left_on='id',right_on='Id').drop('Id',axis=1)
```

```python
df1.join(df5,lsuffix='l',rsuffix='r',how='outer')
```
