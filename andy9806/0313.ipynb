```python
np.random.seed = 313
ages = np.random.randint(120,size=12)
ages
```

```python
df = pd.DataFrame()
```

```python
df['ages'] = pd.Series(ages)
df['cats'] = pd.Series(cats)
df
```

```python
# df.cats.fillna('미분류')
```

```python
df['cats'] = df.cats.astype('object')  #데이터타입이 'object' 여야 astype 가 가능
# df.drop(columns='cat',inplace=True)
df.cats.fillna('미분류',inplace =True)
df
```

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1 # 고객 목록
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2 # 구매 기록
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(left=df2,right=df1)
```

```python
pd.merge(left=df1,right=df2)
```

```python
pd.merge(left=df1,right=df2,on='고객번호')
```

```python
pd.merge(left=df1,right=df2,on='고객번호',how='inner')
# inner join : (기준되는 열) 양쪽 데이터프레임에 모두 속하는 원소들만 남김
```

```python
pd.merge(left=df1,right=df2,on='고객번호',how='left') #왼 쪽만
```

```python
pd.merge(left=df1,right=df2,on='고객번호',how='right') #오른 쪽만
```

```python
pd.merge(left=df1,right=df2,on='고객번호',how='outer') #합집합으로 합치겠다
```
