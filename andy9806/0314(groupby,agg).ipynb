```python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2
```

```python
df2.groupby('key1').count()
```

```python
df2.groupby('key1').mean()
```

```python
df2.groupby('key1').median()
```

```python
df2.groupby('key1').max()
```

```python
df2.groupby('key1').sum()
```

```python
df2.groupby('key1').prod()  #일괄적으로 곱함
```

```python
df2.groupby('key1').std()
```

```python
df2.groupby('key1').var()
```

```python
df2.groupby('key1').quantile(0.25)
```

```python
df2.groupby(['key1','key2']).count()
```

```python
df1.groupby('도시')[['인구']].sum()
```

```python
df1.groupby(['지역','도시'])[['인구']].sum()
```

```python
df2.data1.groupby([df2.key1,df2.key2]).sum()
```

```python
df1.groupby([df1.지역,df1.연도]).sum()
```

```python
df1.groupby([df1.지역,df1.연도]).sum().unstack()
```

```python
df1.groupby([df1.지역,df1.연도]).sum().unstack('연도')
```
