```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.title('Line plot')
plt.plot([i**2 for i in range(15)],marker='D') #1개의 매개변수 >> y축에 표시될 데이터 ,x축은 index
plt.show()
```

```python
plt.title('X tick')
plt.plot(np.arange(10,50,10),[i**2 for i in range(4)]) #x,y도 가능 but 개수 동일해야함
plt.show()
```

```python
def draw_line(grid=False,style='b-'):
  plt.plot(np.arange(10,50,10),[i**2 for i in range(4)],style)
  plt.grid(grid)
  plt.show()
```

```python
draw_line(grid=True)
```

```python
X = np.linspace(-np.pi,np.pi,100)
C,S = np.cos(X),np.sin(X)
```

```python
plt.plot(X,C,'r',label='cos')
plt.plot(X,S,'g',label='sin')
plt.legend()
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(np.arange(5)**2)
plt.subplot(2,1,2)
plt.plot(np.arange(10),np.arange(10)**3)
plt.tight_layout()
plt.show()
```

```python
y=[2,3,1]
x=np.arange(3)
plt.xticks(x,['ga','na','da'])
# plt.yticks(sorted(y))
plt.bar(x,y)
plt.show()
```

```python
y=[2,3,1]
x=np.arange(3)
plt.yticks(x,['ga','na','da'])
plt.xticks(sorted(y))
plt.barh(x,y)
plt.show()
```

```python
labels=['호박','사과','정어리','엄마손']
sizes = [15,30,45,10]
colors = ['yellowgreen','gold','skyblue','coral']
explode = (0,0.1,0,0.2)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.show();
```
